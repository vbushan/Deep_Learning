# Part 1



Import the required libraries-

```python
import numpy as np
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense,Dropout,Flatten
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.metrics import mean_squared_error
from tensorflow.keras import initializers
from matplotlib import pyplot as plt
from keras.models import Model
from sklearn.manifold import TSNE as tsne
from sklearn.decomposition import PCA as pca
import librosa
import IPython.display as ipd
```

Import the mnist dataset.

```python
mnist = tf.keras.datasets.mnist
data=mnist.load_data()
```

Extract the training and test sets from the dataset.

```python
train_x=data[0][0]
train_y=data[0][1].reshape(-1,1)
test_x=data[1][0]
test_y=data[1][1].reshape(-1,1)
```

Process the datasets - Scale the pixel values to range (0,1), and one-hot encode the labels.

```python
# Process the data

def process_data(x,y):


  # Flatten the data
  samples=[]
  for sample in x[:]:
    samples.append(sample.flatten())

  x=np.array(samples)

  # Encode the labels
  enc=OneHotEncoder()

  enc.fit(y)

  onehotlabels = enc.transform(y).toarray()
  y=onehotlabels

  # Shuffle the data
  x,y=shuffle(x,y)

  # Scale the data
  scaler=MinMaxScaler(feature_range=(0,1))

  x=scaler.fit_transform(x)


  return x,y



train_x,train_y=process_data(train_x,train_y)
test_x,test_y=process_data(test_x,test_y)
```

Configure the GPU.

```python
# Configure GPU

physical_devices=tf.config.experimental.list_physical_devices("GPU")
tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

Build a baseline model with 5 hidden layers- each with 1024 units. Compile and train the model.

```python
baseline_model=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu',kernel_initializer=initializers.HeNormal),
    Dense(units=1024,activation='relu',kernel_initializer=initializers.HeNormal),
    Dense(units=1024,activation='relu',kernel_initializer=initializers.HeNormal),
    Dense(units=1024,activation='relu',kernel_initializer=initializers.HeNormal),
    Dense(units=1024,activation='relu',kernel_initializer=initializers.HeNormal),
    
    Dense(units=10,activation='softmax')

    ]
)

baseline_model.compile(optimizer=Adam(),loss='binary_crossentropy',metrics=['accuracy']);
baseline_model_hist=baseline_model.fit(x=train_x,y=train_y,batch_size=1024,epochs=50,shuffle=True,verbose=2,validation_data=(test_x,test_y))
```

Generate predictions using the baseline model and compute the baseline validation accuracy- 

```python
baseline_weights=baseline_model.get_weights()
last_layer_weights=(baseline_weights[-2],baseline_weights[-1])
```

```python
predictions=baseline_model.predict(x=test_x,batch_size=32,verbose=0)
rounded_predictions=np.argmax(predictions,axis=-1)
base_line_accuracy=sum(np.argmax(test_y,axis=-1)==rounded_predictions)/len(rounded_predictions)
print(f'Baseline Test accuracy= {base_line_accuracy}')
```

Write helper functions to compute the SVD components of the baseline layer weights- 

```python
def svd(weights):
  s,u,v=tf.linalg.svd(weights,compute_uv=True,full_matrices=True)
  return s,u,v


svd_components=[]
for i in range(5):
  svd_components.append(svd(baseline_weights[i*2]))
```

```python
def compute_weights(layer_num,num):

  baseline_layer_weights=baseline_weights[2*layer_num]
  baseline_layer_bias=baseline_weights[2*layer_num+1]


  #s,u,v=tf.linalg.svd(baseline_layer_weights,compute_uv=True,full_matrices=True)
  s,u,v=svd_components[layer_num]
  s=tf.linalg.diag(s)

  r_weights=tf.tensordot(tf.tensordot(u[:,:num],s[:num,:num],axes=1),tf.transpose(v[:,:num]),axes=1)

  return (r_weights,baseline_layer_bias)
```

Design models which use- 10, 20, 50, 100, 500, full (all) singular values of the layer weights. Use the helper functions executed above to compute and load the weights in each model.

```python
model_d_10=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    
    Dense(units=10,activation='softmax')

    ]
)


for i in range(len(model_d_10.layers)-1):
  model_d_10.layers[i].set_weights(compute_weights(i,10))


model_d_10.layers[-1].set_weights(last_layer_weights)
```

```python
model_d_20=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    
    Dense(units=10,activation='softmax')

    ]
)



for i in range(len(model_d_20.layers)-1):
  model_d_20.layers[i].set_weights(compute_weights(i,20))

model_d_20.layers[-1].set_weights(last_layer_weights)
```

```python
model_d_50=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    
    Dense(units=10,activation='softmax')

    ]
)


for i in range(len(model_d_50.layers)-1):
  model_d_50.layers[i].set_weights(compute_weights(i,50))

model_d_50.layers[-1].set_weights(last_layer_weights)
```

```python
model_d_100=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    
    Dense(units=10,activation='softmax')

    ]
)


for i in range(len(model_d_100.layers)-1):
  model_d_100.layers[i].set_weights(compute_weights(i,100))

model_d_100.layers[-1].set_weights(last_layer_weights)
```

```python
model_d_500=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    
    Dense(units=10,activation='softmax')

    ]
)


for i in range(len(model_d_500.layers)-1):
  model_d_500.layers[i].set_weights(compute_weights(i,500))

model_d_500.layers[-1].set_weights(last_layer_weights)
```

```python
model_full=Sequential(
    [
    
    Dense(units=1024,input_shape=(784,),activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    Dense(units=1024,activation='relu'),
    
    Dense(units=10,activation='softmax')

    ]
)



model_full.layers[0].set_weights(compute_weights(0,784))
model_full.layers[1].set_weights(compute_weights(1,1024))
model_full.layers[2].set_weights(compute_weights(2,1024))
model_full.layers[3].set_weights(compute_weights(3,1024))
model_full.layers[4].set_weights(compute_weights(4,1024))
model_full.layers[-1].set_weights(last_layer_weights)
```

Compute and plot the validation accuracies of each of these models.

```python
def compute_test_accuracy(model):
  
  predictions=model.predict(x=test_x,batch_size=32,verbose=0)
  rounded_predictions=np.argmax(predictions,axis=-1)

  return sum(np.argmax(test_y,axis=-1)==rounded_predictions)/len(rounded_predictions)
```

```python
result=[]
result.append(compute_test_accuracy(model_d_10))
result.append(compute_test_accuracy(model_d_20))
result.append(compute_test_accuracy(model_d_50))
result.append(compute_test_accuracy(model_d_100))
result.append(compute_test_accuracy(model_d_500))
result.append(compute_test_accuracy(model_full))
```

```python
print(result)
```

```python
from matplotlib import pyplot as plt

plt.plot(result, label='Compressed Models');
plt.title('Accuracy plot');
plt.xlabel('Model Number');
plt.ylabel('Test accuracy');
plt.scatter([6],[base_line_accuracy],label='Baseline Model');
plt.legend();
```

Above we can see that the validation accuracies increase as we increase the value of D, which is the number of singular value components used from the SVD of baseline model weights. The validation accuracy is close enough to baseline model for D>=50. For model with D=500, the validation accuracy is equal to the baseline model.  

Here we have demonstarted the tradeoff between weight compression and model accuracy. A moderate number of singular value components is enough to emulate the performance of the baseline model.

Build a model similar to the baseline model above. So, the model below will have 2(5)+1=11 layers in total.

```python
model_d_20=Sequential(
    [
    
    Dense(units=20,input_shape=(784,),activation=None,use_bias=False,name='layer_1_u'),
    Dense(units=1024,activation='relu',name='layer_1_v'),
    

    Dense(units=20,activation=None,use_bias=False,name='layer_2_u'),
    Dense(units=1024,activation='relu',name='layer_2_v'),
     

    Dense(units=20,activation=None,use_bias=False,name='layer_3_u'),
    Dense(units=1024,activation='relu',name='layer_3_v'),
     
    Dense(units=20,activation=None,use_bias=False,name='layer_4_u'),
    Dense(units=1024,activation='relu',name='layer_4_v'),
     
    Dense(units=20,activation=None,use_bias=False,name='layer_5_u'),
    Dense(units=1024,activation='relu',name='layer_5_v'),
     
    
    
    Dense(units=10,activation='softmax',name='last_layer')

    ]
)

```

Decompose the layer weights in the baseline model and load them into the model above.

```python
for i in range(len(svd_components)):
  u_20=svd_components[i][1][:,:20]
  s_20=tf.linalg.diag(svd_components[i][0])[:20,:20]
  v=tf.transpose(svd_components[i][2][:,:20])

  v_20=tf.linalg.tensordot(s_20,v,axes=1)


  #print(u_20.shape, s_20.shape,v_20.shape)
  model_d_20.layers[2*i].set_weights((u_20,))
  
  model_d_20.layers[2*i+1].set_weights((v_20,baseline_weights[2*i+1]))


model_d_20.layers[-1].set_weights(last_layer_weights)
```

# Part 3

Now, let's see if we can make the model aware of the loss due to SVD decomposition of weights, and achieve better performance.

Design a model with custom layers such that the for every forward pass we decompose the original weights and compute the activations.

```python
def svd(weights):
  s,u,v=tf.linalg.svd(weights,compute_uv=True,full_matrices=True)
  return s,u,v

@tf.custom_gradient
def custom_op(w):
    #print('In OP')
    s,u,v = svd(w) # do SVD
    s=tf.linalg.diag(s)

    s_20=s[:20,:20]
    u_20=u[:,:20]
    v_20=tf.transpose(v[:,:20])
    compr_w=tf.linalg.tensordot(tf.linalg.tensordot(u_20,s_20,axes=1),v_20,axes=1)


    def custom_grad(dy):
        #print('In grad')
        #grad = tf.ones(shape=w.shape) # compute gradient
        return dy
    
    return compr_w, custom_grad


class CustomLayer(keras.layers.Layer):

    def __init__(self,units,layer_num,name,input_shape):
      super(CustomLayer, self).__init__()
      self.units=units
      self.w=tf.Variable(
            initial_value=baseline_weights[2*layer_num],
            trainable=True,
        )
      self.b=tf.Variable(
            initial_value=baseline_weights[2*layer_num+1],
            trainable=True,
        )
      
      
        
    def call(self, x):

        compr_w=custom_op(self.w)  
        a=keras.activations.relu(tf.matmul(x,compr_w)+self.b)
        return a
 
```

Build the model and load weights of the baseline model using the layer number.

```python
model=Sequential(
    [
     
     CustomLayer(units=1024,input_shape=(784,),layer_num=0,name='layer_1'),
     CustomLayer(units=1024,input_shape=(1024,),layer_num=1,name='layer_2'),
     CustomLayer(units=1024,input_shape=(1024,),layer_num=2,name='layer_3'),
     CustomLayer(units=1024,input_shape=(1024,),layer_num=3,name='layer_4'),
     CustomLayer(units=1024,input_shape=(1024,),layer_num=4,name='layer_5'),

     Dense(units=10,input_shape=(1024,),activation='softmax',name='last_layer',weights=[last_layer_weights[0],last_layer_weights[1]])
    ]
)
```

```python
model.compile(optimizer=Adam(),loss='binary_crossentropy',metrics=['accuracy']);
```

```python
model_hist=model.fit(x=train_x,y=train_y,batch_size=1024,epochs=10,shuffle=True,verbose=2,validation_data=(test_x,test_y))
```

```python
model.summary()
```

As we can see from the output, we could achieve resonable validation accuracy by making the model aware of the decomposition procedure.

First lets compute the complex and magintude spectrum of the audio files in the training and validation aset.

```python
from concurrent.futures import ProcessPoolExecutor
import numpy as np
from tqdm import tqdm

def process_data(path,files):

  spect=[]
  mag_spect=[]

  for aud in tqdm(files):
    s, sr=librosa.load(path+aud, sr=None)
    S=librosa.stft(s, n_fft=1024, hop_length=512)
    
    spect.append(S)
    
    mod_S=np.abs(S)
    mag_spect.append(mod_S)

    del s,S,mod_S
  
  return spect,mag_spect
  
```

Compute and plot the dsitribution of SNRs of denoised audios in the training and validation set.

```python
import librosa
import tqdm
import numpy as np


def evaluate(org_path, rec_path,keyword=None):
  
  result=[]

  def comp_snr(org,rec):

    if len(org)<len(rec):
      rec=rec[:len(org)]
    
    elif len(org)>len(rec):
      org=org[:len(rec)]

    
    num=1e-20

    for i in range(len(org)):
      num+=org[i]**2

    denom=1e-20

    for i in range(len(org)):
      diff=org[i]-rec[i]
      denom+=diff**2
    
    return 10*np.log10(num/denom)


  org_files=[file for file in sorted(os.listdir(org_path)) if file.startswith(keyword)]

  rec_files=sorted(os.listdir(rec_path),key=lambda x: int(x.split('.')[0].split('_')[1]))

  for org_file,rec_file in tqdm.tqdm(zip(sorted(org_files),rec_files)):
    org,_=librosa.load(org_path+org_file, sr=None)
    rec,_=librosa.load(rec_path+rec_file, sr=None)

    snr_value=comp_snr(org,rec)
    print(snr_value)
    result.append(snr_value)
  
  return result


org_path=path+'tr/'
rec_path=path+'train_denoise/'

train_snrs=evaluate(org_path,rec_path,'trs')

```

```python
org_path=path+'v/'
rec_path=path+'val_denoise/'

val_snrs=evaluate(org_path,rec_path,'vs')
```

```python
from matplotlib import pyplot as plt

def plot_histogram(arr):
    plt.figure(figsize=[10,8])
    plt.hist(x=arr, bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('SNRs',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.title('SNR distribution',fontsize=15)
    plt.show()
    
plot_histogram(train_snrs)
```

```python
plot_histogram(val_snrs)
```

As we can see, our model performed decently on both the sets- Highest SNR>=35 in the training set, while highest SNR>=25 in the validation set

```python
import numpy as np
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import mean_squared_error
from matplotlib import pyplot as plt
import IPython.display as ipd
```

```python
mnist = tf.keras.datasets.mnist
data=mnist.load_data()
```

```python
train_x=data[0][0]
train_y=data[0][1].reshape(-1,1)
```

```python
# Process the data

def clean_data(x,y):


  # Flatten the data
  samples=[]
  for sample in x[:]:
    samples.append(sample.flatten())

  x=np.array(samples)

  # Encode the labels
  enc=OneHotEncoder()

  enc.fit(y)

  onehotlabels = enc.transform(y).toarray()
  y=onehotlabels

  # Shuffle the data
  x,y=shuffle(x,y)

  # Scale the data
  scaler=MinMaxScaler(feature_range=(0,1))

  x=scaler.fit_transform(x)

  print(x.shape)

  final_x=[]
  for i in range(len(x)):
    final_x.append(x[i].reshape(28,28))

  final_x=np.array(final_x)
  
  print(final_x.shape)


  return final_x,y



train_x,train_y=clean_data(train_x,train_y)
```

```python
from tqdm import tqdm


def div_frames(data):

  processed_data=[]
  for n in tqdm(range(len(data))):
    sample=[]
    for i in range(0,data[n].shape[0],7):
      for j in range(0,data[n].shape[1],7):
        
        sample.append(data[n][i:i+7,j:j+7].flatten())
    
    sample=np.array(sample)
    processed_data.append(sample)
    del sample

  return np.array(processed_data)


train_x=div_frames(train_x)
```

```python
train_x.shape,train_y.shape
```

```python
def preprocess(data):

  x=[]
  y=[]
  for i in tqdm(range(len(data))):
    x.append(data[i][:-1,:])
    y.append(data[i][1:,:])

  x=np.array(x)
  y=np.array(y)

  return x,y

x,y=preprocess(train_x)
```

```python
model=Sequential([
                
                  Input(shape=(None,49)),
                  LSTM(activation='tanh',units=64,return_sequences=True,name='layer_1'),
                  LSTM(activation='tanh',units=64,return_sequences=True,name='layer_2'),
                  Dense(activation='sigmoid',units=49,name='final_layer',kernel_initializer=keras.initializers.HeNormal)
                
                ])
```

```python
model.compile(optimizer=Adam(),loss=mean_squared_error,metrics=['mean_squared_error']);
```

```python
model.summary()
```

```python
model_hist=model.fit(x=x,y=y,batch_size=1024,epochs=100,shuffle=True,verbose=2, validation_split=0.1)
```

```python
def predict(sample):
  sample=sample[tf.newaxis,:,:]

  for i in range(8):
    prediction=model.predict(sample,batch_size=1)
    sample=np.append(sample,prediction[0][-1].reshape((1,49))[tf.newaxis,:,:],axis=1)


  pred_image=sample[0]

  recon_image=np.zeros((28,28))

  seq=0
  for i in range(0,28,7):
    for j in range(0,28,7):
        
      recon_image[i:i+7,j:j+7]=pred_image[seq].reshape(7,7)
      seq+=1

  return recon_image
```

```python
def evaluate(data,labels):

  indexes={i:[] for i in range(10)}

  for i in range(len(labels)):
    key=np.argmax(labels[i])
    if len(indexes[key])<10:
      indexes[key].append(i)
  
  #print(indexes)

  fig, axes=plt.subplots(nrows=10,ncols=10)



  for i in range(100):
    row=i//10
    column=i%10

    sample=data[indexes[row][column]][:8,:]
    predicted_image=predict(sample)

    axes[row,column].imshow(predicted_image)


evaluate(x,train_y)
```
